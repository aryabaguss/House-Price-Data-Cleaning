# House Price EDA

Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Read Dataset

```python
data_path = 'house_price.csv'

df = pd.read_csv(data_path)
df.head()
```

```python
# Sale Price Histogram
sns.histplot(data = df, x = 'SalePrice')
```

```python
# Sale Price Histogram by each MSZoning
sns.histplot(data = df, x = 'SalePrice', hue = 'MSZoning')
```

```python
sns.histplot(data = df, x = 'GrLivArea', hue = 'LotShape')
```

```python
# Box Plot Sale Price
sns.boxplot(data = df, x = 'SalePrice')
```

```python
# Box Plot Ground Living Area
sns.boxplot(data = df, x = 'GrLivArea')
```

```python
# Scatterplot GrLivArea vs SalePrice
sns.scatterplot(data = df, x = 'GrLivArea', y = 'SalePrice')
```

```python
sns.scatterplot(data = df, x = 'LotFrontage', y = 'SalePrice')
```

```python
sns.boxplot(data = df, x = 'SalePrice', y = 'MSZoning', hue = 'MSZoning')
```

```python
sns.boxplot(data = df, x = 'SalePrice', y = 'MSZoning', hue = 'CentralAir')
```

```python
sns.boxplot(data = df, x = 'GrLivArea', y = 'Foundation', hue = 'LandSlope')
```

```python
# Line Plot House Price Average each year
sns.lineplot(data=df, x="YrSold", y="SalePrice")
```

```python
sns.lineplot(data=df, x="YrSold", y="GrLivArea")
```

## Correlation heatmap

```python
# Calculate correlation value between numeric columns
corr = df.select_dtypes(include = ['number']).corr(method = 'pearson')

# Show correlation value using heatmap
sns.heatmap(corr, cmap = 'magma')
```

```python
df.drop('Id', axis = 1, inplace = True)

# Calculate correlation value between numeric columns
corr = df.select_dtypes(include = ['number']).corr(method = 'pearson')

# Show correlation value using heatmap
sns.heatmap(corr, cmap = 'magma')
```

## Styling Plot

```python
df2 = pd.read_csv(data_path)
```

```python
# Sale Price Boxplot
sns.boxplot(data = df2, x = 'SalePrice')

# Scaling X Ticks
plt.xticks([0,100000,500000,1000000]
           , labels=['0', '100K', '500K', '1M'])

# Show BoxPlot
plt.show()
```

```python
# Line Plot average house price each year
sns.lineplot(data = df2, x = 'YrSold', y = 'SalePrice')

# Scale the xticks
plt.xticks([2006, 2007, 2008, 2009, 2010])

# Label X
plt.xlabel('Year')

# Label Y
plt.ylabel('Sale Price')

# Show BoxPlot
plt.show()
```

```python
# Pivot Data to shows Sale Price on each year for each zone
piv_data = df2.pivot_table(index="YrSold", columns = "MSZoning", values="SalePrice")
piv_data
```

```python
# Create Line Plot
sns.lineplot(data = piv_data)

# Scale the xticks
plt.xticks([2006, 2007, 2008, 2009, 2010])

# Label X
plt.xlabel('Year')

# Label Y
plt.ylabel('Sale Price')

# Showing Plot
plt.show()
```

```python
# Histogram Plot
sns.histplot(data = df2, x = 'SalePrice')

# Adding Ttile to Plot
plt.title("Histogram of SalePrice")

# Showing Vertical Line
plt.axvline(df2['SalePrice'].mean(), color = 'r', linestyle = 'dashed', label = 'mean')

# Showing Legend
plt.legend()

# Showing Plot
plt.show()
```

```python
# Histogram Plot
sns.histplot(data = df2, x = 'SalePrice')

# Adding Ttile to Plot
plt.title("Histogram of SalePrice")

# Showing Vertical Line
plt.axvline(df2['SalePrice'].median(), color = 'r', linestyle = 'dashed', label = 'median')

# Showing Legend
plt.legend()

# Showing Plot
plt.show()
```

```python
# Plot Size 8x4
plt.figure(figsize = (8,4))

# Create boxplot
sns.boxplot(data = df2, x = "SalePrice")
plt.xticks([0,100000,500000,1000000]
           , labels=['0', '100K', '500K', '1M'])

# Showing Plot
plt.show()
```
